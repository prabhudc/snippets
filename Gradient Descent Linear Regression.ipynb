# Hand-coded Gradient Descent for Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random X values
X = np.random.randn(1000,1)
# Create dummy-target variable y
y = 5 + 2*X + 2*np.random.randn(1000,1)
# For the sake calculation , concatenate 1s for the intercept position
X_b = np.c_[np.ones((1000, 1)), X]

# Initialize a random theta
theta = np.random.randn(2,1)
# Initialize a random learning-rate. Try other values
eta = .1
thetas = []
# 100 iterations of gradient descent seem to be good-enough
for i in range(100):
    gradients = (eta * (1/X_b.shape[0])*X_b.T.dot(y - X_b.dot(theta)))
    theta = theta + gradients
    # Record values in an array for plotting
    if i%10 == 0:
        thetas.append(theta)
```

```python
plt.scatter(X,y,c='b',alpha=.2,marker='.',)
for i in range(10):
    X_linear = np.linspace(min(X),max(X),len(X))
    y_hat = thetas[i][0] + thetas[i][1]*X_linear
    plt.plot(X_linear,y_hat,c='red',alpha=i*.7)
plt.show()
```
